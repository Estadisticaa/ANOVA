```python
import matplotlib.pyplot as plt
import numpy as np
import random
from mpl_toolkits.mplot3d import Axes3D
```

## Escenario 1

El concepto de interacci√≥n es posible representarlo

```python
def f(x1,x2):
    return 35.5 + 10.5*x1 + 5.5*x2 + 17.5*x1*x2

x = np.linspace(-1,1,100)
p = np.linspace(-1,1,100)

X,P = np.meshgrid(x,p)
y = f(X,P)
```

```python
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')

ax.plot_surface(X, P, y, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none');
```

```python
cs = plt.contour(X,P,y,15)
plt.clabel(cs, inline=1, fontsize=8)

plt.show()
```
